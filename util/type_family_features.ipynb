```python
import pandas as pd
import json
```

```python
def read_file(filename):
    types_data = pd.read_csv(filename,sep='\s+')
#    types_data['Type'] = types_data['Type'].map(lambda x: x.lstrip('dbo:'))
#    types_data['Parent'] = types_data['Parent'].map(lambda x: x.lstrip('dbo:'))
    return types_data

#save features as a dictionary format in file

def save_features(features, filename):
    with open(filename,'w') as f:
        json.dump(features,f)
    f.close()
    return

#load features as dictionary
def load_features(filename):
    try:
        with open(filename,'r') as f:
            features = json.load(f)
        return features 
    except:
        return None
```

```python
filename = './data/dbpedia_types.tsv'

tdata = read_file(filename)
tdata.head()
```

```python
def family_features(data):
    features = {}
    for i,t in enumerate(data['Type']):
        features[t] = {}
        features[t]['depth'] = int(data.iloc[i]['Depth'])
        features[t]['siblings'] = list(data.loc[data.iloc[i]['Parent'] == data['Parent']]['Type'].values)
        features[t]['children'] = list(data.loc[data.iloc[i]['Type'] == data['Parent']]['Type'].values)
    return features


        
```

```python
features = family_features(tdata)
features
```

```python
save_features(features,'./data/type_hierarchy_features.json')
```
